```python
# what to merge: earnings, total_teams from otcap_carreer_earnings.
# re-order and group.

# Merge the DataFrames
merged_df = pd.merge(pro_df, earnings_df, on='Player', how='outer', indicator=True)

# Display where matches weren't found in either DataFrame
no_match = merged_df[merged_df['_merge'] != 'both']

# Display the merged DataFrame and no_match DataFrame
display("Merged DataFrame:")

display(merged_df.info())
# display("\nRows with no match:")
# display(no_match)
```

```python
display(merged_df.head())
display(merged_df.tail())
merged_df.columns
```

```python
# Convert specified columns to integers
import numpy as np
combined_df = merged_df.copy()


# # Replace 9999, pd.NA, and '<NA>' with np.nan across the entire DataFrame
combined_df.replace([9999, pd.NA, '<NA>', 'NaN', 'nan'], np.nan, inplace=True)

display('after replacing 9999 and NA and <NA> and nan: ')
display(combined_df.head())
display(combined_df.tail())
```

```python
import re
import numpy as np
# Display the "Earnings By Team Dict" column
print(combined_df['Earnings By Team Dict'])

# Convert all values in 'Earnings By Team Dict' to strings
combined_df['Earnings By Team Dict'] = combined_df['Earnings By Team Dict'].astype(str)

# Function to parse the string and split into teams and earnings
def split_earnings(earnings_str):
    teams = re.findall(r"'(.*?)'", earnings_str)
    earnings = re.findall(r": ([\d.]+)", earnings_str)
    earnings = [float(earning) for earning in earnings]
    return pd.Series([teams, earnings])

# Apply the function to create the new columns
combined_df[['Teams_List', 'Earnings_List']] = combined_df['Earnings By Team Dict'].apply(split_earnings)


# Display the updated DataFrame
display(combined_df.head())
display(combined_df.tail())
```

```python
# investigate and organize columns, then determine which to drop.
# Slicing columns 0 to 9 (first 10 columns, next 10, etc.
df_1 = combined_df.iloc[:, :10]
df_2 = combined_df.iloc[:, 10:20]
df_3 = combined_df.iloc[:, 20:30]

display(df_1.head())
display(df_2.head())
display(df_3.head())
combined_df.columns
```

```python
# data type assigning with to_numeric; change % strings to Float64
display('before conversions:: ')
display(nfl_clean_df.info())

# Function to convert percentage strings to integers
def convert_to_int(series):
    series = series.astype(str).str.rstrip('%')  # Ensure all entries are strings and remove the '%' sign
    series = pd.to_numeric(series, errors='coerce')  # Convert to numeric (float)
    series = series.fillna(0).astype(int)  # Fill NaN with 0 and convert to int
    return series

nfl_clean_df['comp_%'] = convert_to_int(nfl_clean_df['comp_%'])
nfl_clean_df['int_%'] = convert_to_int(nfl_clean_df['int_%'])

display(nfl_clean_df.info())
display(nfl_clean_df.tail())
display(nfl_clean_df.columns)
```
