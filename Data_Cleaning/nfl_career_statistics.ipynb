### NEW DATAFRAME  cleanING_df

```python
import numpy as np

cleaning_df = merged_df.copy()

# Replace values of 9999 with np.nan
# There are no 'nan' values at this point.
cleaning_df.replace([9999], np.nan, inplace=True)

float_to_int = ['pass_yds', 'comp', 'att', 'TD', 'int', 'sacks', 'pick_6', 
                'pass_yds_game', 'GWD', '4QC', 'num_teams', 'hof_yes', 
                'yrs_played', 'drafted', 'final']

# Safely Convert the specified columns to Int64:
# NOTE: Converting the columns to Int64 converts the NaN to the <NA> format.
#       If you convert the <NA> back to NaN, the rest of the columns will be converted to Float64

# Sorting Int64 columns with <NA> values works smoothly in Pandas. <NA> values will be placed 
#   at the end when sorting in ascending order, and at the beginning when sorting in descending order

for column in float_to_int:
    if column in cleaning_df.columns:
        cleaning_df[column] = pd.to_numeric(cleaning_df[column], errors='coerce').astype('Int64')


display(cleaning_df.iloc[:,:15].head(2))
display(cleaning_df.iloc[:,15:30].head(2))
display(cleaning_df.iloc[:,:15].tail(2))
display(cleaning_df.iloc[:,15:30].tail(2))
```

```python
# ver 2: Break out "Earnings By Team Dict" into teams and earnings, 
# cast to pd.Series in the Function.
import ast

def split_dict_to_columns(d_str):
    '''Function splits the string representation of the dictionary into two lists
      and places them in new columns teams, team_earnings:
      {'Packers': 100.697, 'Vikings': 28.0, 'Jets': 36.706}
        ['Packers', 'Vikings', 'Jets' ]   
        [100.697, 28.0, 36.706}    '''
    if pd.notna(d_str):  # Check if the value is not NaN
        try:
            d = ast.literal_eval(d_str)
            teams = list(d.keys())
            team_earnings = list(d.values())
            return pd.Series([teams, team_earnings])
        except (ValueError, SyntaxError) as e:
            print(f"Error parsing: {d_str}")
            return pd.Series([[], []])
    else:
        return pd.Series([[], []])

# Apply the function to the 'earn_by_team' column
cleaning_df[['teams', 'team_earnings']] = cleaning_df['earn_by_team'].apply(split_dict_to_columns)

display(cleaning_df.head())
display(cleaning_df.tail())

# pd.Series
# NOTE: casting to a pd.Series is helpful when you want to leverage Pandas' powerful functionalities. 
# pd.Series makes it easier to work with data in a tabular format, enabling quick operations like adding 
# it to a DataFrame, performing vectorized operations, or utilizing Pandas-specific methods.

# use pd.Series when you need data in a format that can be seamlessly integrated into Pandas operations. 
# Itâ€™s particularly useful when:
#  + to leverage Pandasâ€™ vectorized operations (fast computations over series).
#  + you need compatibility with DataFrames for further manipulation.
#  + you want easy access to Pandasâ€™ methods for a single column of data.
```

```python
# change comp and int % columns so they can have numerical analysis:
# data type assigning with to_numeric; change % strings to Float64

# Function to convert percentage strings to floats
def convert_to_float(series):
    series = series.astype(str).str.rstrip('%')  # Ensure all entries are strings and remove the '%' sign
    series = pd.to_numeric(series, errors='coerce')  # Convert to numeric (float)
    return series

# Apply function to the 'comp_%' and 'int_%' columns
cleaning_df['comp_%'] = convert_to_float(cleaning_df['comp_%'])
cleaning_df['int_%'] = convert_to_float(cleaning_df['int_%'])

display(cleaning_df.info())
display(cleaning_df.tail())
display(cleaning_df.columns)
```
