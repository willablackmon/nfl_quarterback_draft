```python
# Get a list of college players without NFL data and NFL data without college data
# Do an outer merge, indicator  = True


merged_df = pd.merge(cfb_df, label_df, left_on='player', right_on='player_nfl', how='outer', indicator=True)

both_df = merged_df[merged_df['_merge'] == 'both']
display(f'both: {len(both_df)}')

display(both_df.iloc[:,:20].head())
display(both_df.iloc[:,20:].head())
```
