### NEW DATAFRAME   sorted_cleaning_df
Prior to running compares of duplicate columns and replacing them using the below function, 
let's make a 'cleaning' version (copy) of the sorted_df 

```python
# CHECK for lost values and totals in sorted_df before creating new DF. 
display(sorted_df.info())
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_colwidth', None)

# display(sorted_df[['player', 'team', 'all_star']].sort_values(by='player'))
```

```python
# compares pairs of columns, get best value, and drop extraneous columns.
sorted_cleaning_df = sorted_df.copy()

# Define pairs as a list of tuples [(col_1, col_2), ...]
pairs = [
    ('comp_%_x', 'comp_%_y'),
    ('drafted_x', 'drafted_y'),
    ('hof_x', 'hof_y'),
    ('pass_yards', 'pass_yds'),
    ('TD_x', 'TD_y'),
    ('int_x', 'int_y'),
    ('wAV_x', 'wAV_y')
]

def compare_columns(df, pairs, drop=False):
    '''Function compares pairs of columns and creates::
        1) new Boolean column stating if the values in the columns are _equal
        2) best value is placed in new column called _combine (combined from col_1, col_2)

    + NaN values are considered 'equal.  <NA> and <NA> are considered equal.
    + col_1 val is used in _combine column if both col_1 and col_2 are 'valid'
    + If there is no resolution when determining which col to put in _combine, use NaN
    '''
    # Initialize new DataFrame for results, insert 'player' column, for reference
    result_df = pd.DataFrame()
    result_df['player'] = df['player']
    
    for col_1, col_2 in pairs:
        # new column names, using the first 3 letters of col_1
        equal_column_name = f"{col_1[:3]}_equal"
        combine_column_name = f"{col_1[:3]}_combine"
        
        # Cols to be compared 
        result_df[col_1] = df[col_1]
        result_df[col_2] = df[col_2]
        
        # Create the combine column based on the specified logic
        result_df[combine_column_name] = np.where(
            df[col_1].notna() & (df[col_1] != ''),  # If col_1 is valid
            df[col_1],
            np.where(
                df[col_2].notna() & (df[col_2] != ''),  # If col_2 is valid
                df[col_2],
                np.nan  # If neither are valid
            )
        )
        
        # Ensure the combine column has the same data type as the columns being compared
        result_df[combine_column_name] = result_df[combine_column_name].astype(df[col_1].dtype)

        # Create the equal column: True if both values are equal or both are NaN
        result_df[equal_column_name] = (df[col_1] == df[col_2]) | (df[col_1].isna() & df[col_2].isna())
        
        # If both values are NaN, ensure the result is True
        result_df[equal_column_name] = result_df[equal_column_name].fillna(False)

    # Optionally drop the original columns

    # Optionally drop the original columns
    if drop:
        for col_1, col_2 in pairs:
            result_df.drop([col_1, col_2], axis=1, inplace=True, errors='ignore')
            print(f'Drop={drop}; dropping cols: {col_1}, {col_2}')
    
    return result_df

# set drop to True if you wish to go ahead and drop the compared columns after creating the combined col _combine and 
#  the _equal column.
sorted_compare_df = compare_columns(sorted_cleaning_df, pairs, drop=True)
display(sorted_compare_df.head())
```

```python
bool_cols = [col for col in sorted_compare_df.columns if '_equal' in col]
print(bool_cols)
```

```python
# Check for "bad" values in the different combined columns:

# Get all combine columns
combine_cols = [col for col in sorted_compare_df.columns if '_combine' in col]
print(bool_cols)

# Generalized function to check for invalid values
def invalid_col_values(value):
    # Let NaN values pass as valid
    if pd.isna(value):
        return False  # NaN is considered valid
    if isinstance(value, str):
        # Consider 'bad' if it's an empty string or contains unwanted characters
        return value == '' or 'bad' in value
    return False  # Assume other types are valid

# Loop through each combine column, check for invalid values, and create individual DataFrames
for combine_col in combine_cols:
    # Create mask to identify rows with invalid values in the column (excluding NaN)
    invalid_mask = sorted_compare_df[combine_col].apply(invalid_col_values)

    # Filter the DataFrame to show rows with invalid values (excluding NaN)
    filtered_invalid_rows = sorted_compare_df[invalid_mask][['player', combine_col]]

    # Create the column prefix from the combine column name
    column_prefix = combine_col[:3]

    # Select the 'player' column and related columns with the same prefix
    columns_to_show = ['player'] + [col for col in sorted_compare_df.columns if col.startswith(column_prefix)]
    
    # Include only invalid rows (non-NaN)
    filtered_invalid_rows = sorted_compare_df[invalid_mask][columns_to_show]

    # Display the column name
    print(f"({len(filtered_invalid_rows)}) invalid values in column: {combine_col}. (There will be NaN.)")
    
    # Display the filtered DataFrame, even if it's empty
    display(filtered_invalid_rows.head())

```
