```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import display


# Test URLs and requests...
url_1 = 'https://www.pro-football-reference.com/leaders/pass_yds_career.htm'
url_2 = 'https://www.pro-football-reference.com/leaders/pass_yds_career.htm'
# GET request to the URL
response = requests.get(url_1)
tables = {}
# Check if the request was successful
if response.status_code == 200:
    # Parse HTML content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    # Find all tables in the HTML content, print the number found.
    tables = soup.find_all('table')
    print(f"Number of tables found: {len(tables)}")

    # Iterate over all tables and print a snippet to identify the correct one
    for i, table in enumerate(tables):
        print(f"\nTable {i}:")
        print(table.prettify()[:1000])  # Print snippet of table
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")

```

```python
url_list = [ 'https://www.pro-football-reference.com/leaders/pass_yds_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_cmp_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_att_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_td_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_rating_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_int_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_sacked_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_sacked_yds_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pick_six_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_yds_per_g_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_cmp_perc_career.htm'
            ,'https://www.pro-football-reference.com/leaders/pass_int_perc_career.htm'
            ,'https://www.pro-football-reference.com/leaders/gwd_career.htm'
            ,'https://www.pro-football-reference.com/leaders/comebacks_career.htm'
            ,'https://www.pro-football-reference.com/leaders/career_av_career.htm'
]

# Dictionary to store each DataFrame
dataframes = {}

for url in url_list:
    response = requests.get(url)
    if response.status_code == 200:
        name = url.split("leaders/")[1].replace('_career.htm', '')
        print(f"Let's cook with {name}")
        # Parse HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        tables = soup.find_all('table')
        # Find all tables in the HTML content, print the number found
        if tables:
            print(f"{len(tables)} tables found.")
            table_index = 0  # Replace with the correct table index after identifying
            df = pd.read_html(str(tables[table_index]), header=0)[0]
            # df = df.rename(columns={'Player': 'Player', 'Tm': f'Tm_{name}', 'Years': f'Years_{name}'})
            
            #  save each DataFrame into a dictionary called dataframes, with name as the key
            dataframes[name] = df  # Save each DataFrame with a unique name
        else:
            print("No tables found.")
    else:
        print(f"Failed to fetch the webpage. Status code: {response.status_code}")

# Display the DataFrames
for name, df in dataframes.items():
    print(f"\nDataFrame: {name}")
    display(df.head())
```

```python
# Process each DataFrame and put in dataframes dictionary of all dataframes
#  with name of dataframe as key, and the Dataframe as the value.

for name, df in dataframes.items():
    print(f'{name} {df.columns}')
    if 'Rank' in df.columns:
        df.drop(columns=['Rank'], inplace=True)
    print(f"\nDataFrame: {name}")
    # display(df.head())
    # df.columns

# Assuming dataframes is your dictionary with keys as DataFrame names and values as DataFrames
for name, df in dataframes.items():
    print(f"\nDataFrame: {name}")
    print("Columns:", df.columns.tolist())
```

```python
game_winning_df = dataframes['gwd'].drop(columns=['Link'])
game_winning_df.head()
dataframes['game_win_drives'] = game_winning_df
display(game_winning_df.head())
del dataframes['gwd']

comebacks_df = dataframes['comebacks'].drop(columns=['Link'])
display(comebacks_df.head())
dataframes['comebacks'] = comebacks_df
```

```python
# List of DataFrame names
df_names = list(dataframes.keys())
# insert the first dataframe:
merged_df = dataframes[df_names[0]]

# interatively merge the Dataframes in df_names:
for name in df_names[1:]:
    if name in dataframes:
        merged_df = pd.merge(merged_df, dataframes[name], 
                             on=['Player', 'Years', 'Tm'], how='outer')
        display(f'{name}')
        display(merged_df.head())
        display(merged_df.info())
```

```python
# map old column names to new more column names
merged_df.columns
old_cols = ['Player', 'Yds_x', 'Years', 'Tm', 'Cmp', 'Att', 'TD', 'Rate', 'Int', \
       'Sk', 'Yds_y', 'Pick6', 'Y/G', 'Cmp%', 'Int%', '4QC', 'GWD']
new_cols = ['Player', 'pass_yds', 'yrs_active', 'team', 'comp', 'att', 'pass_td', 'passer_rate', 'int', \
       'sacks', 'yds_lost', 'pick_6', 'pass_yds_game', 'comp_%', 'int_%', 'comebacks', 'game_win_drive']

# Create lists into dictionary
columns_mapping = dict(zip(old_cols, new_cols))
career_stats_df = merged_df.rename(columns= columns_mapping)
display(career_stats_df.head())
career_stats_df.columns
```

```python
stats_clean_df = career_stats_df.copy()

# split the team column into num_teams and teams:
def split_team(team):
    '''Function to split 'team' into 'num_teams' and 'teams'''
    if 'TM' in team:
        return pd.Series([int(team[0]), 'tm'])
    else:
        return pd.Series([1, team.upper()])

# Apply the function and create new columns
stats_clean_df[['num_teams', 'teams']] = stats_clean_df['team'].apply(split_team)
# Drop the 'team' column
career_stats_df = career_stats_df.drop(columns=['team'])


# Add 'hof' column: 1 if 'Player' contains '+', else 0
stats_clean_df['hof'] = stats_clean_df['Player'].apply(lambda x: 1 if '+' in x else 0)
# Strip the '+' from the 'Player' column values
stats_clean_df['Player'] = stats_clean_df['Player'].str.replace('+', '', regex=False)


# Display the updated DataFrame
display(stats_clean_df.head())
```

```python
#  count total years played from years_active and create a new column to capture it

# Function to calculate total years played
def calculate_years_played(yrs_active):
    start_year, end_year = yrs_active.split('-')
    start_year = int(start_year)
    end_year = int(end_year) if int(end_year) != 9999 else 2023
    return end_year - start_year + 1

# Apply the function and create a new column 'total_years_played'
stats_clean_df['tot_yrs_played'] = stats_clean_df['yrs_active'].apply(calculate_years_played)

# Display the updated DataFrame
display(stats_clean_df.head())
```

```python
# Convert specified columns to integers
import numpy as np

# Convert specified columns to numeric, retaining NaNs
columns_to_convert = ['comp', 'att', 'pass_td', 'int', 'sacks', 'pick_6', 'comebacks', 'game_win_drive']
stats_clean_df[columns_to_convert] = stats_clean_df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Force conversion to integer, handling NaNs properly
stats_clean_df[columns_to_convert] = stats_clean_df[columns_to_convert].astype(pd.Int64Dtype())


display(stats_clean_df.head())
display(stats_clean_df.info())
```
