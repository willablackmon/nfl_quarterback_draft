```python
# merge (prospects_df - player_name, profiles_df - player_name_prof) using fuzzy wuzzy, to allow for some possible differences in the way the name is spelled. 
# print an indicator when there is NOT a match

# Method:
# 1. Create new column 'player_name_match' in dataframe 1 prospects_df
# For each value of player_name, search 'player_name_prof' in dataframe 2: profiles_df using process.extractOne to get the best match (or None if no good match).
#  this best match gets added to 'player_name_match' in dataframe 1 prospects_df
# 2. Then, merge dataframe 1 prospects_df (on 'player_name_match') and dataframe 2 profiles_df (on 'player_name_prof')
#   Maintain the original 'player_name' in dataframe 1 for reference.

# pip install fuzzywuzzy
# Uses slow pure-python SequenceMatcher;  Install python-Levenshtein an optional dependency that speeds things 
# pip install python-Levenshtein

from fuzzywuzzy import process

# Find the best match for each player name in qb_profiles_clean_df
def get_best_match(name, choices, threshold=80):
#  process.extractOne returns tuple w/ 3 elements: match, score, and the index.
    match, score, _ = process.extractOne(name, choices) 
    return match if score > threshold else None

# Create the 'player_name_match' column
prospects_df['player_name_match'] = prospects_df['player_name'].apply(
    lambda x: get_best_match(x, profiles_df['player_name_prof'])
)

# Merge the dataframes
merged_df = pd.merge(
    prospects_df,
    profiles_df,
    left_on='player_name_match',
    right_on='player_name_prof',
    how='outer',
    indicator=True
)

display(merged_df.head())
display(merged_df.tail())
```

```python
compare_cols = {
    'height': 'height_prof',
    'player_id': 'player_id_prof',#'alt_player_id_prof',
    'grade': 'grade_prof',
    'guid': 'guid_prof', 
    'ovr_rk': 'ovr_rk_prof', 
    'player_name': 'player_name_prof', 
    'pos_abbr': 'pos_abbr_prof',
    'pos_rk': 'pos_rk_prof', 
    'school': 'school_prof', 
    'weight': 'weight_prof'
}

compare_df = merged_df.copy()

def compare_columns(row, col1, col2):
    '''This function checks two columns, pulling from a dictionary of columns to compare.
    It creates a new column called "{col1}_alt".
      If values are equal or both NaN, place value from col1 in col1_alt
      If there is a valid value in both columns, place value from col2 in col1_alt with an appended *.
      If one value is NaN, and the other is a valid value, return the valid value.'''
    
    val1, val2 = row[col1], row[col2]
    if (pd.isna(val1) and pd.isna(val2)) or (val1 == val2):
        return val1
    elif pd.notna(val1) and pd.notna(val2):
        if isinstance(val1, float) and isinstance(val2, float):
            return val1 if len(str(val1)) >= len(str(val2)) else val2
        else:
            return f'{val1}*'
    elif pd.notna(val1) and pd.isna(val2):
        return val1
    elif pd.isna(val1) and pd.notna(val2):
        return val2
    else:
        return 'none found'

for col1, col2 in compare_cols.items():
    # Check if columns exist in DataFrame
    if col1 in compare_df.columns and col2 in compare_df.columns:
        # Creating a new column to store comparison results
        compare_df[f'{col1}_compare'] = compare_df.apply(lambda row: compare_columns(row, col1, col2), axis=1)
    else:
        print(f"Columns {col1} or {col2} not found in DataFrame")

# DataFrame of compared columns
compare_cols_list = [col1 for col1 in compare_cols] + [col2 for col2 in compare_cols.values()] + [f'{col1}_compare' for col1 in compare_cols]
df_col_compares = compare_df[compare_cols_list]

# Show mismatch rows
mismatch_df = df_col_compares[df_col_compares.apply(lambda x: 'none found' in x.values, axis=1)]

display(mismatch_df.head())
display(mismatch_df.info())
display(df_col_compares[pd.notna(df_col_compares[[f'{col1}_compare' for col1 in compare_cols]].values).any(axis=1)].head())
display(df_col_compares.info())
```
