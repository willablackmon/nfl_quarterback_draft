```python
# url_1 = 'https://www.sports-reference.com/cfb/leaders/pass-cmp-player-career.html'
# https://www.sports-reference.com/cfb/leaders/pass-att-player-career.html
# https://www.sports-reference.com/cfb/leaders/pass-cmp-pct-player-career.html
# https://www.sports-reference.com/cfb/leaders/adj-pass-yds-per-att-player-career.html
# https://www.sports-reference.com/cfb/leaders/pass-yds-per-att-player-career.html
# https://www.sports-reference.com/cfb/leaders/pass-rating-player-career.html
# https://www.sports-reference.com/cfb/leaders/pass-int-player-career.html
# https://www.sports-reference.com/cfb/leaders/pass-yds-per-cmp-player-career.html

base_url = "https://www.sports-reference.com"
url_suffix = "-player-career.html"

cfb_qb_data = {
# passing
    "Pass Completions": "/cfb/leaders/pass-cmp",
    "Pass Attempts": "/cfb/leaders/pass-att",
    "Pass Completion Percentage": "/cfb/leaders/pass-cmp-pct",
    "Passing Yards": "/cfb/leaders/pass-yds",
    "Passing Yards Per Attempt": "/cfb/leaders/pass-yds-per-att",
    "Adjusted Passing Yards Per Attempt": "/cfb/leaders/adj-pass-yds-per-att",
    "Passing Touchdowns": "/cfb/leaders/pass-td",
    "Passing Interceptions": "/cfb/leaders/pass-int",
    "Passing Efficiency Rating": "/cfb/leaders/pass-rating",
# passing
    "Passing Yards Per Completion": "/cfb/leaders/pass-yds-per-cmp",
    "Yards Gained Per Game Played": "/cfb/leaders/pass-yds-per-game",
# total offense
    "Total Plays": "/cfb/leaders/tot-plays",
    "Total Yards": "/cfb/leaders/tot-yds",
    "Total Yards Per Play": "/cfb/leaders/tot-yds-per-play",
    "TD Responsible For": "/cfb/leaders/tot-td"    }
```

```python
import time

# Dictionary to store each DataFrame
dataframes = {}

for name, path in cfb_qb_data.items():
    url = f'{base_url}{path}{url_suffix}'
    time.sleep(1)
    response = requests.get(url)
    if response.status_code == 200:
        print(f"Loading table for {name} from {url}")

# Parse HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')
        tables = soup.find_all('table')
# Find all tables in the HTML content, print the number found
        if tables:
            print(f"{len(tables)} tables found.")
            table_index = 0  # Replace with the correct table index after identifying
            df = pd.read_html(str(tables[table_index]), header=0)[0]
            dataframes[name] = df  # Save each DataFrame with a unique name
        else:
            print("No tables found.")
    else:
        print(f"Failed to fetch {url}. Status code: {response.status_code}")

# Display the DataFrames
for name, df in dataframes.items():
    print(f"\nDataFrame: {name}")
    display(df.head(3))
```

```python
# Process each DataFrame
for name, df in dataframes.items():
    print(f'{name} {df.columns}')
    if 'Rank' in df.columns:
        df.drop(columns=['Rank'], inplace=True)
    print(f"\nDataFrame: {name}")
    # display(df.head())
    # df.columns

# Assuming dataframes is your dictionary with keys as DataFrame names and values as DataFrames
for name, df in dataframes.items():
    print(f"\nDataFrame: {name}")
    print("Columns:", df.columns.tolist())
```

```python
import pandas as pd
import pickle
path = '../Data_Artifacts/'

# Load the dictionary from the pickle file
with open(f'{path}cfb_ref_stats_raw_dataframes.pkl', 'rb') as f:
    dataframes_loaded = pickle.load(f)

# Verify by printing one of the DataFrames
display(dataframes_loaded['Passing Touchdowns'])  

dataframes = dataframes_loaded.copy()
```

```python
# for all tables (except the first one), drop Rank, From, To columns

#  Columns From and To are redundant across all tables.  Should be able 
#   to disambiguate the names by using their School also.  
#   We'll check for dupes at the end.


# enumerate provides an index (i) with the item (name, df), allowing skipping first item w/o manually tracking indices:
for i, (name, df) in enumerate(dataframes.items()):
    if i > 0:  # Skip the first DataFrame
        df.drop(['Rk', 'From', 'To'], axis=1, inplace=True)
        print(f'{name}: {df.columns}')
```

```python
# Merge all of the Tables on the Player Name, School

# Get list of DataFrame names
df_names = list(dataframes.keys())
merged_df = dataframes[df_names[0]]

# iteratively merge the Dataframes in df_names:
for name in df_names[1:]:
    if name in dataframes:
        # if name == 'career_av':
        #     print('skipping df career_av')
        #     continue
        merged_df = pd.merge(merged_df, dataframes[name], 
                             on=['Player', 'School'], how='outer')  #outer 
        display(f'{name}')
        display(merged_df.head(3))
```

```python
# Find any rows where 'Player' is duplicated
duplicates = merged_df[merged_df.duplicated(subset=['Player'], keep=False)].sort_values('Player')
display(duplicates)
display(merged_df.columns)
```

```python
# Remove the * from the Player column; drop Rank columns

# replace can replace characters anywhere in the string
# rstrip only removes characters from the end
# Both must be used with the str accessor, which is necessary for string operations on pandas Series
# merged_df['Player'] = merged_df['Player'].replace('*', '')
merged_df['Player'] = merged_df['Player'].str.rstrip('*')

# Drop Rank column, remove the * from all the player names
merged_df.drop(columns='Rk', inplace=True)

display(merged_df.columns)
display(merged_df.head())
```

```python
# change To and From to int64, fill an na values
# calc total years played and place in new col 'yr

# Fill NaN values with 0 or another appropriate value
merged_df['From'] = merged_df['From'].fillna(0).astype('int64')
merged_df['To'] = merged_df['To'].fillna(0).astype('int64')

# Adding the yrs_played column
merged_df['yrs_played'] = merged_df['To'] - merged_df['From']

display(merged_df.head())
display(merged_df.columns)
```

```python
# map old column names to new more column names
display(merged_df.columns)

cols_update = {'Player' : 'player',  
               'From'   : 'from',               
               'To'     : 'to',             
               'yrs_played': 'yrs_played',  
               'School' : 'school',           
               'Rate'   : 'passer_eff_rate',   # Passing Efficiency Rating     
               'Att'    : 'att',               # Pass Attempts
               'Cmp'    : 'comp',              # Pass Completions
               'Pct'    : 'comp_%',            # Pass Completion Percentage
               'Yds_x'  : 'pass_yds',           # Passing Yards
               'TD_x'   : 'pass_td',            # Passing Touchdowns
               'Int'    : 'int',                # Passing Interceptions
                'Y/A'   : 'pyds_att',           # Passing Yards Per Attempt
                'AY/A'  : 'adj_pyds_att',       # Adjusted Passing Yards Per Attempt
                'Y/C'   : 'pass-yds-per-cmp',   # Passing Yards Per Completion     
                'Y/G'   : 'yds_pgame',          # Yards Gained Per Game Played  
                'Plays' : 'tot_plays',          # Total Plays    
                'Yds_y' : 'tot_yds',            # 
                'Avg'   : 'yds_pplay',          # Average Yards Per Play         
                'TD_y'  : 'tot_td'              # TD Responsible For
}

merged_df = merged_df.rename(columns= cols_update)
merged_df.head()
```
