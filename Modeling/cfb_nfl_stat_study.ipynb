### 1) Investigate Data

```python
display(df.info())
display(df.iloc[:,:20].head())
display(df.iloc[:,20:].head())
display(df.columns)
```

```python
df['nfl_success'].value_counts()
```

```python
# start with College Football statistical data and the aggregate y target data "nfl_success"

y = df['nfl_success']
X = df[['passer_eff_rate', 'yrs_played', 
       'comp', 'att', 'comp_%', 'pass_yds', 'pass_td', 'int', 'tot_plays',
       'tot_yds', 'tot_td', 'pyds_att', 'adj_pyds_att', 'pass-yds-per-cmp',
       'yds_pgame', 'yds_pplay']]
```

```python
display(X.head())
display(X.columns)
```

```python
nan_counts = X.isna().sum()
print(len(X))
print(nan_counts)
```

### 2) Handle NA Values
+ Fill na with 0.
+ Fill na using SimpleImputer from sklearn

###     a.  Use 0 for the NA values

```python
# replace NaN with 0
X0 = X.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the data into X_train, X_test, y_train, y_test
X0_train, X0_test, y0_train, y0_test = train_test_split(X0, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
X0_scaler = StandardScaler()

# Fit the Standard Scaler with the training data 
X0_scaler.fit(X0_train)

# Scale the TRAINING DATA and TEST DATA SEPARATELY with the Transform Function
X0_train_scaled = X0_scaler.transform(X0_train)
X0_test_scaled = X0_scaler.transform(X0_test)
```

###     b. use SimpleImputer to creare a second set of data to fill in the NA values (mean strategy)

```python
from sklearn.impute import SimpleImputer

# Create SimpleImputer with mean strategy
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the DataFrame and transform it
Xi = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

print(Xi[:10])
```

### 3) Compute Correlation Matrix: On the cleaned DataFrame

```python
import matplotlib as plt
import seaborn as sns
import numpy as np
```

```python
# Calculate correlation matrix of X data (0 for NA replacement):
X0_corr_matrix = X0.corr().abs().round(2)

# Plot heatmaps
plt.figure(figsize=(12, 4))  # Adjust the width and height as needed
sns.heatmap(X0_corr_matrix, annot=True, cmap='coolwarm')
f = plt.figure()
plt.show()

# Select upper triangle of correlation matrix
upper = X0_corr_matrix.where(np.triu(np.ones(X0_corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.8
high_corr = [column for column in upper.columns if any(upper[column] > 0.8)]
display(f'highly correlated data (0 for NA replacement): {high_corr}')
```

```python
# Calculate correlation matrix of X data (imputer (mean) for NA replacement):
Xi_corr_matrix = Xi.corr().abs().round(2)

# Plot heatmaps
plt.figure(figsize=(12, 4))  # Adjust the width and height as needed
sns.heatmap(Xi_corr_matrix, annot=True, cmap='coolwarm')
f = plt.figure()
plt.show()

# Select upper triangle of correlation matrix
upper = Xi_corr_matrix.where(np.triu(np.ones(Xi_corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.8
high_corr = [column for column in upper.columns if any(upper[column] > 0.8)]
display(f'highly correlated data (imputer (mean) for NA replacement): {high_corr}')
```

### 4) Train-Test Split

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split the data into training and testing sets
Xi_train, Xi_test, yi_train, yi_test = train_test_split(Xi, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler instance
Xi_scaler = StandardScaler()
# Fit the Standard Scaler with the training data 
Xi_scaler.fit(Xi_train)
# Scale the TRAINING DATA and TEST DATA SEPARATELY with the Transform Function
Xi_train_scaled = Xi_scaler.transform(Xi_train)
Xi_test_scaled = Xi_scaler.transform(Xi_test)
```

### Create and Fit a Logistic Regression Model
+ create a Logistic Regression model
+ fit/training it to the training data, 
+ make predictions with the testing data,
Analyze by viewing model's accuracy score.

### LR Model With the 0 for NA data:

```python
from sklearn.linear_model import LogisticRegression

# Create a LogisticRegression instance
lr0_model = LogisticRegression(random_state = 13)
# Fit the LogisticRegression with the training data 
lr0_model.fit(X0_train_scaled, y0_train)
# Check fit of training data...
print(f"LogReg Model training/fit score (with 0 for NA data): {lr0_model.score(X0_train_scaled, y0_train):.4f}")
```

```python
# Calc predictions  and calc accuracy core with y0_test (actual) vs. y_predicted
y0_preds = lr0_model.predict(X0_test_scaled)
acc_score = accuracy_score(y0_test, y0_preds)
print(f'LogReg Model accuracy_score (with 0 for NA data): {acc_score:.4f}')
```

### LR Model with the imputer (mean) data for NA

```python
# Create a LogisticRegression instance
lri_model = LogisticRegression(random_state = 13)
# Fit the LogisticRegression with the training data 
lri_model.fit(Xi_train_scaled, yi_train)
# Check fit of training data...
print(f"LogReg Model training/fit score (with imputer for NA data ): {lri_model.score(Xi_train_scaled, yi_train):.4f}")
```

```python
# Calc predictions and calc accuracy core with y0_test (actual) vs. y_predicted
yi_preds = lri_model.predict(Xi_test_scaled)
acc_score = accuracy_score(yi_test, yi_preds)
print(f'LogReg imputer data accuracy_score: {acc_score:.4f}')

# same value, one line.
score = lri_model.score(Xi_test_scaled, yi_test)
print(f'LogReg Model score (imputer data for na): {score:.4f}')
```

## Create and Fit a Random Forest Classifier Model

### Random Forest Model with the 0 for NA data:

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
# Train the Random Forest model
rfc0 = RandomForestClassifier(random_state=13, n_estimators=500). \
                    fit(X0_train_scaled, y0_train)
# Evaluate the Training model
print(f'Random Forest Model training/fit Score (0 for NA data): {rfc0.score(X0_train_scaled, y0_train):.4f}')
```

```python
# Make y predictions and calc the accuracy score by evaluating the y predictions vs. the y_test values
y0_preds = rfc0.predict(X0_test_scaled)
acc_score0 = accuracy_score(y0_test, y0_preds)
print(f'Random Forest Model accuracy score (0 for NA data) accuracy_score: {acc_score0:.4f}')
```

### Random Forest Model (with the inputer for NA data):

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
# Train the Random Forest model
rfci = RandomForestClassifier(random_state=13, n_estimators=500). \
                    fit(Xi_train_scaled, yi_train)
# Evaluate the Training model
print(f'Random Forest Model training/fit Score (imputer (mean) for NA data): {rfci.score(Xi_train_scaled, yi_train):.4f}')
```

```python
# Make y predictions and calc the accuracy score by evaluating the y predictions vs. the y_test values
yi_preds = rfci.predict(Xi_test_scaled)
acc_scorei = accuracy_score(yi_test, yi_preds)  # actual, predicted
print(f'Random Forest Model accuracy score (imputer (mean) for NA data): {acc_scorei:.4f}')
```

### Investigate Feature Imporances for the Random Forest trained on 0s for NA and the one trained on Imputer for NA

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Investigate Feature Imporances for the Random Forest trained on 0s for NA
# rfc0 is your trained Random Forest model and X0 is your DataFrame
importances_0 = np.round(rfc0.feature_importances_ * 100, 3)  # Get importances in %
feature_names_0 = X0.columns  # Get feature names

# Create a DataFrame for feature importances
X0_feat_import_df = pd.DataFrame({'Feature': feature_names_0, 'Importance': importances_0})

# Display sorted feature importances
display(X0_feat_import_df.sort_values(by='Importance', ascending=False))

# Visualize with Seaborn bar chart of Feature Importance
plt.figure(figsize=(8, 4))  # Width=8, Height=4
bar_plot = sns.barplot(x='Importance', y='Feature', data=X0_feat_import_df, palette='viridis')
plt.xlabel('Importance (%)')
plt.ylabel('Feature')
plt.title('X Feature Importances')
plt.show()
```

```python
# Investigate Feature Imporances for the Random Forest trained on Imputer for NA
# rfci is your trained Random Forest model and X0 is your DataFrame
importances_i = np.round(rfci.feature_importances_ * 100, 3)  # Get importances in %
feature_names_i = Xi.columns  # Get feature names

# Create a DataFrame for feature importances
Xi_feat_import_df = pd.DataFrame({'Feature': feature_names_i, 'Importance': importances_i})

# Display sorted feature importances
display(Xi_feat_import_df.sort_values(by='Importance', ascending=False))

# Visualize with Seaborn bar chart of Feature Importance
plt.figure(figsize=(8, 4))  # Width=8, Height=4
bar_plot = sns.barplot(x='Importance', y='Feature', data=Xi_feat_import_df, palette='viridis')
plt.xlabel('Importance (%)')
plt.ylabel('Feature')
plt.title('X Feature Importances')
plt.show()
```
